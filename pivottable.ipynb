```python
#pivot table
import random
import numpy as np
import pandas as pd
fruit=['apple', 'orange', 'banana']*3
item=['a','b','c','d','e','f','g','g','g']
price=np.random.randint(1,10,9)
df=pd.DataFrame({'fruit':fruit, 'item':item, 'price':price})
df
```

```python
df.groupby('fruit').agg({'price':'sum'})
```

```python
df.groupby(['fruit', 'item']).agg({'price':'sum'})
```

```python
df.pivot_table(values='price', index='fruit', columns='item', aggfunc='sum', fill_value=0)
```
