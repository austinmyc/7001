### 1. File i/o

```python
#store text file by line to list
with open('./data_in/mbox_short.txt', encoding='utf-8') as handle:
    x=[]
    for i in handle:
        x.append(i.strip())
        
#write with string/list
with open('./data_out/cheatsheet.txt', 'w') as outhandle:
    outhandle.write(x[1])
    outhandle.writelines(x)
```

```python
#store csv by line to list
import csv
with open('./data_in/io_csv.csv', encoding='utf-8') as handle:
    reader=csv.reader(handle, delimiter=',')
    x=[]
    for row in reader:
        x.append(row)

#write csv file
with open('./data_out/cheatsheet2.csv','w',newline='') as outhandle:
    wr=csv.writer(outhandle)
    wr.writerow(['name1', 'age1'])
    wr.writerows(x)
```

```python
#BeautifulSoup
import requests
import pandas as pd
from bs4 import BeautifulSoup

requests.packages.urllib3.disable_warnings()
url = 'http://books.toscrape.com/'
headers = {
'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.71 Safari/537.36'}
book = requests.get(url, headers = headers, verify=False).text

soup=BeautifulSoup(book, 'html.parser')
soup.find_all('div',{'class':'image_container'})[:3]
```

### 6. Others

```python
#Filter
students = [
    {"name": "Hermione", "house": "Gryffindor"},
    {"name": "Harry", "house": "Gryffindor"},
    {"name": "Ron", "house": "Gryffindor"},
    {"name": "Draco", "house": "Slytherin"},
]

gryffindors = filter(lambda s: s["house"] == "Gryffindor", students)

for gryffindor in sorted(gryffindors, key=lambda s: s["name"]):
    print(gryffindor["name"])
```

```python
#Dictionary comprehension
students = ["Hermione", "Harry", "Ron"]

gryffindors = {student: "Gryffindor" for student in students}

print(gryffindors)
```

```python
#Dictionary unpacking
def total(galleons, sickles, knuts):
    return (galleons * 17 + sickles) * 29 + knuts


coins = {"galleons": 100, "sickles": 50, "knuts": 25}

print(total(**coins), "Knuts")
```
